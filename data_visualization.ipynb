### What is Data visualization?
Data visualization is the graphical or visual representation of data and information using elements like charts, graphs, maps, and infographics to make complex data easier to understand and interpret.

By translating numerical or textual data into visual formats, data visualization helps people quickly spot patterns, trends, correlations, and outliers, supporting better, faster, and more informed decision-making.

```python
# Defining data for the dataframe
data = {
    'Basket': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O', 'P'],
    'Apples': [10, 20, 30, 56, 40, 40, 67, 47, 40, 4, 49, 52, 5, 56, 35, 45],
    'Bananas': [15, 6, 3, 45, 67, 44, 45, 11, 14, 18, 13, 12, 1, 34, 12, 12]
}

# Creating the dataframe
df = pd.DataFrame(data)

df
```

### Visualization in IDE

Create graphs and visualizations that match your chosen color scheme.

```python
# Calculate the sums
sum_apples = df['Apples'].sum()
sum_bananas = df['Bananas'].sum()

# Create a bar chart
plt.bar(['Apples', 'Bananas'], [sum_apples, sum_bananas], color=['red', 'blue'])

# Set a title
plt.title('Comparison of total Apples and Bananas')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ---------------------------------------------------------
# 1. DATA PREPARATION
# ---------------------------------------------------------
# Small subset of Gapminder data for demonstration
# (GDP, Life Exp, Population, Continent Color)
gdp_cap = [974.58, 5937.03, 6223.37, 4797.23, 12779.38, 34435.37, 36126.49, 29796.05, 1391.25, 33692.61, 1441.28, 2452.21, 4952.40]
life_exp = [43.82, 76.42, 72.30, 42.73, 75.32, 81.24, 79.83, 75.64, 64.06, 79.44, 62.29, 65.55, 72.96]
pop = [31.89, 3.60, 33.33, 12.42, 40.30, 20.43, 8.20, 12.92, 10.10, 4.60, 12.45, 1100.0, 1300.0]
col = ['red', 'green', 'blue', 'blue', 'yellow', 'black', 'green', 'yellow', 'red', 'green', 'blue', 'blue', 'red']

# Note: The last two data points in the lists above represent India and China
# so that your labels appear near actual dots.

# ---------------------------------------------------------
# 2. PLOTTING CODE (Your Final Code)
# ---------------------------------------------------------

# Scatter plot
# We use np.array(pop) so we can multiply the whole list by 2 at once
plt.scatter(x = gdp_cap, y = life_exp, s = np.array(pop) * 2, c = col, alpha = 0.8)

# Previous customizations
plt.xscale('log')
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000], ['1k','10k','100k'])

# ---------------------------------------------------------
# 3. FINAL ADDITIONS (Text & Grid)
# ---------------------------------------------------------

# Additional customizations
# These coordinates (x, y) place the text on the plot
plt.text(1550, 71, 'India')
plt.text(5700, 80, 'China')

# Add grid() call
plt.grid(True)

# Show the plot
plt.show()
```
